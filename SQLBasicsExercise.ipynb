SQL Basics Exercise (Practice)

```python
# imports
import pandas as pd
from sqlalchemy import create_engine
```

```python
#data
school = pd.read_csv('/content/School_Details.csv')
school.head()
```

```python
school.rename(columns = {'DistrictHQDistance (in Km)':'DistrictHQDistance', 'CRCDistance (in Km)':'CRCDistance',
                              'BACDistance (in Km)':'BACDistance'}, inplace = True)
school.head()
```

```python
# Create SQL engine
engine = create_engine('sqlite:///school.db', echo=True)
# Connection
conn = engine.connect()
# add the dataframe as a table in sqlite
sqlite_table = "school_details"
school.to_sql(sqlite_table, conn, if_exists='fail')
```

```python
sql = """
SELECT EstablishmentName, SchoolLevel, Website
FROM school_details
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName, EstablishmentName
FROM school_details
WHERE DistrictHQDistance > 100
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Write your SQL query in this string
sql = """
SELECT EstablishmentName, SchoolLevel, DistrictHQDistance
FROM school_details
WHERE SchoolLevel = 'JHS'
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName, SchoolLevel, DistrictHQDistance
FROM school_details
WHERE EstablishmentName LIKE '%JUNIOR HIGH%'
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
# Write your SQL query in this string
sql = """
SELECT EstablishmentName, SchoolLevel, DistrictHQDistance
FROM school_details
WHERE EstablishmentName LIKE 'C%'
ORDER BY EstablishmentName 
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName
FROM school_details
WHERE LocatedInRuralOrUrban LIKE '%Urban%'
ORDER BY EstablishmentName 
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```

```python
sql = """
SELECT EstablishmentName AS name
FROM school_details
WHERE LocatedInRuralOrUrban LIKE '%Urban%'
ORDER BY EstablishmentName 
"""
# Read the sql query into pandas to view the results
df = pd.read_sql_query(sql, engine)
df.head()
```
